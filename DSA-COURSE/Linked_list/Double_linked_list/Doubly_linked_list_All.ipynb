```python
class Node:
    def __init__(self, data) -> None:
        self.data = data
        self.next = None
        self.previous = None

class Doubly_Linked_list:
    def __init__(self) -> None:
        self.head = None
        self.tail = None
        self.count = 0

    def is_empty(self):
        return (self.head is None) or (self.count == 0)
    
    def insert_at_first(self, data):
        newnode = Node(data)
        if self.is_empty():
            self.head = newnode
            self.tail = newnode
        else:
            newnode.next = self.head
            self.head.previous = newnode
            self.head = newnode
        self.count += 1

    def insert_at_last(self, data):
        newnode = Node(data)
        if self.is_empty():
            self.head = newnode
            self.tail = newnode
        else:
            newnode.previous = self.tail
            self.tail.next = newnode
            self.tail = newnode
        self.count += 1

    def search(self, e):
        if not self.is_empty():
            x = self.head
            while x:
                if (x.data == e):
                    return x
                x = x.next
            return None
    

    def insert_at_after_value(self, data, e):
        if self.is_empty():
            return None
        else:
            x = self.search(e)
            newnode = Node(data)
            newnode.next = x.next
            newnode.previous = x
            newnode.next.previous = newnode
            x.next = newnode     

    def delete_at_first(self):
        if self.is_empty():
            return None
        else:
            self.head = self.head.next
            self.head.previous = None
        self.count -= 1

    def delete_at_Last(self):
        if self.is_empty():
            return None
        else:
            self.tail = self.tail.previous
            self.tail.next = None
        self.count -= 1

    def delete_at_after_value(self, e):
        if self.is_empty():
            return None
        elif self.count == 1: 
            self.head = None
            self.tail = None
        else:
            n = self.search(e)
            n.previous.next = n.next
            n.next.previous = n.previous.previous
        self.count -= 1
            
    def show(self):
        if not self.is_empty():
            x = self.head
            while x:
                print(x.data)
                x = x.next
            else:
                return None
            

Dlist = Doubly_Linked_list()
Dlist.is_empty()
Dlist.insert_at_first(111)
Dlist.insert_at_first(222)
Dlist.insert_at_first(333)
Dlist.insert_at_first(787)
Dlist.insert_at_first(398)
Dlist.insert_at_first(5620)
Dlist.insert_at_last(444)
Dlist.insert_at_last(555)
Dlist.insert_at_last(666)
Dlist.insert_at_last(777)
Dlist.insert_at_after_value(12313, 555)
Dlist.delete_at_first()
Dlist.delete_at_Last()
Dlist.delete_at_after_value(555)
Dlist.search(222)
Dlist.show()
```
